```python
def age_checker(age = 0):
    if age<0 or age >150:
        print("Invalid age",age)
    else:
        if age< 18:
            print("You cannot vote")
        elif age > 60:
            print("close to retirement age")   
        else:
            print("you are young") 
```

```python
age = int(input("Enter a valid age:"))
age_checker(age)
```

```python
def canvote(is_citizen, age= 0):
    if not is_citizen:
        print('cannot vote',age)
        return False
    if age >= 18:
        print('can vote at the age',age)
        return True
    else:
        print('cannot vote at the age',age)
        return False

print(canvote(True, 132), '\n')
print(canvote(False, age=31), '\n')
```

```python
#problem - 8 solution.....

class Student:
    count = 0

    def __init__(self,id,name):
        self.id = id
        self.name = name
        self.cgpa = 0.00
        print("Created a Student object named",self.name)

    def get_cgpa(self):
        return  self.cgpa

    def update_cgpa(self,new_cgpa):
        self.cgpa = new_cgpa

  

    def __str__(self):
        s = self.name + " "  + str(self.cgpa)
        return s 
```

```python
n = int(input())
students={}
for i in range(n):
    id = int(input("ID: "))
    name = input("Name: ")
    student = Student(id,name)
    students[id] = student
```

```python
for id in students:
    print(id,students[id])
```

```python
M = int(input("M: "))
for i in range(M):
    id = int(input("Id: "))
    cgpa = float(input("Cgpa: "))
    if id in students:
        students[id].update_cgpa(cgpa)



# solution of practice problem - 8 here.....
```
