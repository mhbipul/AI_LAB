#copy a list


```python
foo = [1,2,3,4,5]
bar = foo
bar.append(6)
print(foo,bar,'\n')
```

```python
foo = [1,2,3,4,5]
bar = foo.copy()
bar.append(7)
print(bar)
```

```python
import copy
foo = [1,2,3,4,5,[]]
bar = copy.deepcopy(foo)
bar[-1].append(7)
print(foo,bar,'\n')
```

##Dictionary

```python
d = {
    "name": "Mahmudul Hasan Bipul",
     "email" : "mbipul201025@bscse.uiu.ac.bd",
     "joined":"2020 spring",
  
}
print(d)
print(d["name"])
d["name"]="hasan"
print(d)
```

```python
d = {
    "name": "Mahmudul Hasan Bipul",
     "email" : "mbipul201025@bscse.uiu.ac.bd",
     "joined":"2020 spring",
      "position": "student",
      3  : "???",
     1 : "what?",
     "info": [1,3,"blue"]

}
print(d)
print(d[1])
print(d["info"][2])
print(d.keys())
print(d.values())
```

```python
for key in d.keys():
    print(key,d[key])
```
