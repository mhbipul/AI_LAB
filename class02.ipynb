```python
class Student:
    count = 0
    def __init__(self,id,name) -> None:
        self.name = name #member var
        self.id =name
        self.cgpa = 3.0
        print("Created a Student object named: ",self.name)

    def get_cgpa(self):
        return self.cgpa  
    def update_cgpa(self,new_cgpa):
        self.cgpa = new_cgpa      

    def dummy(): #static function...
        print("This is a dummy function",Student.count)   
        Student.count = Student.count+1 
    def __str__(self):
        s = self.name +" "+ str(self.cgpa)
        return s
```

```python
hasan = Student(25,"Mahmudul Hasan") #  hasan is the object name
print(hasan.get_cgpa())
Student.dummy() #calling the static function.....
Student.dummy()
Student.dummy()
Student.dummy()
Student.dummy()
print(hasan)
```

```python
import math
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def move_left(self): 
         self.x-=1
         return self.x,self.y   
    def move_right(self):
         self.x+=1
         return x,self.y 
    def move_up(self):
        self.y+=1
        return self.x,y 
    def move_down(self):
        self.y-=1  
        return self.x,y 

    def manhattan_distance(p1,p2):
        result = (p1.x - p2.x)+ (p1.y-p1.y)
        return math.abs(result)             
```

```python
p1 = Point(2,4)
p2 = Point(0,0)
#print("Distance : ",Point.manhattan_distance(p1,p2))
print(p1.move_left())
print(p1.move_right())
print(p1.move_up())
print(p1.move_down())
print("For Point2--------")
print(p2.move_left())
print(p2.move_right())
print(p2.move_up())
print(p2.move_down())
```
